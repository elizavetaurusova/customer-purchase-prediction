```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def rename_columns(name):
    new_name = name.replace(' ', '_').lower()
    return new_name
```

```python
df = pd.read_csv(r"D:\Магистратура\archive (5)\superstore_data.csv")
df = df.rename(columns = rename_columns)
```

```python
for col in df.columns:
    plt.figure(figsize=(16, 8))
    x = df['response']
    y = df[col]
    plt.subplots_adjust(bottom=0.25, top=0.79)
    ax = sns.boxplot( x, y)
    plt.xticks(rotation=80)
    plt.draw()
```
